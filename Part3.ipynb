## Function for Visualization 1:

````python

# -- imports --
import matplotlib.pyplot as plt
import json
import matplotlib.colors as mcolors

# global variable, stays the same has 0 chance to change
chapters = list(range(1, 60))  # Chapters 1 to 14

# line chart
def line_chart(entity, data):
    """
    Plots a line chart representing the sentiment of a specified entity over time.

    Parameters:
    entity (str): The character whose sentiment is being analyzed.
    data (list): List of sentiment scores corresponding to each chapter.

    This function creates a line chart with chapters on the x-axis and sentiment scores on the y-axis, 
    displaying how the sentiment of the given entity evolves across chapters.
    """

    plt.figure(figsize=(25, 10))
    plt.plot(chapters, data, marker='o', linestyle='-', color='b')
    plt.title(f'Sentiment of {entity} Over Time')
    plt.xlabel('Chapter')
    plt.ylabel('Sentiment Score')
    plt.grid(True)
    plt.show()
    
    
# line chart for two characters 
def double_line_chart(entity1, entity2, data1, data2):
    """
    Plots a line chart comparing the sentiments of two entities over time.

    Parameters:
    entity1 (str): The first character for comparison.
    entity2 (str): The second character for comparison.
    data1 (list): List of sentiment scores for the first entity corresponding to each chapter.
    data2 (list): List of sentiment scores for the second entity corresponding to each chapter.

    This function creates a comparative line chart for two entities, showing their sentiment trends across chapters.
    """
    plt.figure(figsize=(25, 10))
    
    plt.plot(chapters, data1, marker='o', color='blue', label=entity1)
    plt.plot(chapters, data2, marker='s', color='green', label=entity2)

    plt.title(f'Sentiment of {entity1} vs {entity2} Over Time')
    plt.xlabel('Chapter')
    plt.ylabel('Data Value')
    plt.legend()
    plt.show()
    
    
# bar chart
def bar_chart(entity, data):
    """
    Plots a bar chart representing the sentiment of a specified entity over time.

    Parameters:
    entity (str): The character whose sentiment is being analyzed.
    data (list): List of sentiment scores corresponding to each chapter.

    This function creates a bar chart with chapters on the x-axis and sentiment scores on the y-axis, 
    illustrating the sentiment of the given entity across different chapters.
    """
        
    plt.bar(chapters , data)
    plt.title(f'Sentiment of {entity} Over Time')
    plt.xlabel('Chapter')
    plt.ylabel('Data Value')
    plt.title('Data per Chapter')
    plt.show()

# heatmap
def heatmap(data, row_labels):
    """
    Plots a heatmap representing sentiments across various entities over time.

    Parameters:
    data (2D list): A 2D list of sentiment scores, where each row represents an entity and each column a chapter.
    row_labels (list): List of labels for the rows, each representing an entity.

    This function creates a heatmap to visualize the sentiment scores of various entities across different chapters, 
    using color intensity to represent sentiment magnitude.
    """
    
    fig, ax = plt.subplots(figsize=(25,10))  # Adjust the size as needed

    # Create a colormap
    cmap = mcolors.LinearSegmentedColormap.from_list("", ["blue", "green", "yellow", "red"])

    # Create the heatmap
    cax = ax.matshow(data, cmap=cmap)

    # Set axis labels
    ax.set_xticklabels([] + chapters, rotation=45, ha='left')
    ax.set_yticklabels([] + row_labels)
    
    # Set axis ticks
    ax.set_xticks(range(len(chapters)))
    ax.set_yticks(range(len(row_labels)))


    # Hide the gridlines based on minor ticks
    ax.grid(False)
    ax.grid(which='minor', color='w', linestyle='-', linewidth=2)
    
    # Add a color bar
    plt.colorbar(cax)

    # Add title
    plt.title(f'Everyone\'s Sentiments of  Over Time')

    plt.show()
    

def data_prep(index):
    """
    Prepare and preprocess sentiment data for visualization.

    Args:
    - index (int): The index of the character in the sentiment data list to be visualized.

    Returns:
    list: A list of sentiment values, one for each chapter, with missing values filled with the character's
          average sentiment score throughout the entire book.

    This function reads sentiment data from a JSON file, accesses the data for a specific character
    based on the provided index, and prepares it for visualization. It ensures that sentiment scores are available
    for all chapters, filling in missing values with the character's average sentiment score throughout the entire book.

    After calling the function, sentiment_data will be a list of sentiment values for each chapter.
    """
    
    # prepare the data to avoid NA errors.
    # as placeholder for sentiments of chapters where a character does not occur, we put 
    # their average throughout the entire book
    # this allows the visualization and comparison of characters that are not mentioned in every chapter

    # open sentiment file and load the josn
    with (open('.\\pride_and_prejudice\\pride_and_prejudice_sentiment.json', 'r', encoding="UTF8") as sentiment_file):
        
        # access the data for the character you want to visualize
        character_dict_list = json.load(sentiment_file)
        character = character_dict_list[index]

        
        #add average character rating to every chapter where they are not mentioned (to avoid NA errors when plotting) 
        # find out for which chapters the characters don't have a rating 
        all_chapter_numbers = [str(num) for num in range(1, 60)]  #complete chapter list
        existing_chapter_numbers = character["compound_average_per_chapter"].keys()
        # add the average as values for the missing chapters
        missing_chapter_number = [num for num in all_chapter_numbers if
                                  num not in existing_chapter_numbers]  #all missing chapters

        for chapter in missing_chapter_number:
            character["compound_average_per_chapter"][str(chapter)] = character["compound_average_entire_book"]

        # re-sort (the appending destroys the order) the chapters to assure chronology
        character["compound_average_per_chapter"] = dict(
            sorted(character["compound_average_per_chapter"].items(), key=lambda item: int(item[0])))
        
        # return the prepped data for use
        return  list(character["compound_average_per_chapter"].values())




# the indexes correspond to;
#     0; Elizabeth Bennet   
#     1; Jane Bennet
#     2; Mr Collins
#     3; Mr Bingley
#     4; Mr Darcy
#     5; Lydia Bennet
#     6; lady Catherine de Bourgh

# Example usages for the line and bar plots  
line_chart('Elizabeth Bennet', data_prep(0))
bar_chart('Lydia Bennet', data_prep(5))
double_line_chart("Collins", "Bingley", data_prep(2), data_prep(3))



# example for the heatmap plot
all_character_data = [data_prep(i) for i in range(0, 7)]
character_as_row = ["eliza", "jane", "collins", "darcy", "bingley", "lydia", "lady c"]
heatmap(all_character_data,character_as_row)
````

## Function for Visualization 2:

````python
# visualize grep results for easier comparison
# comarison is limited because grep sentiments are not character specific 

positive_sentiments = ["love", "content", "pleased", "pleasure", "joyful", "glad", "happy", "excited", "hopeful", "loving", "affection", "cheerful","tender", "amiable", "elated", "satisfied", "laugh", "gentle"]
negative_sentiments = ["hate", "hatred", "melancholy", "sorrow", "sad", "disappointed", "regret", "vexed", "rage", "angry", "displeased", "confused"]

# empty list to store the sentiment ratios
ratio_list = []

# prepare grep data;
for i in range(1,60):

    # count positives and negatives per chapter
    positives = 0
    negatives = 0
    
    # open the chapter grep file
    with (open(f".\\pride_and_prejudice\\{i}_Sentiment.txt", "r", encoding="UTF8") as grep_file):
        
        # iterate through the file
        for line in grep_file:
            
            # count the sentiments
            if line.split(":")[0] in positive_sentiments:
                positives += int(line.split(":")[1].strip())
            if line.split(":")[0] in negative_sentiments:
                negatives += int(line.split(":")[1].strip())
    
    # store the sentiment list. The higher the value the more negative the sentence.
    # unfortunately this has to be this way around because dividing by negatives gives zero division errors, so does the inverse value
    ratio_list.append( (negatives / positives))
    
    
# plot the ratios as line chart 
line_chart("grep, the more negative the higher the spike", ratio_list)        
                
        


##
````
